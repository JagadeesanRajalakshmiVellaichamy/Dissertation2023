```python
from googleapiclient.discovery import build
import googleapiclient.errors

#Token and authentication
API_TOKEN = "AIzaSyDPkNg5OIyRc2M3i2eOyR8hMvU5MTH-N7U"
yt = build('youtube', 'v3', developerKey=API_TOKEN)

try:
    #Query build on elections in Tamil which extract video ids to search for comments
    search_response = yt.search().list(
        q='рокро┐ро░родрооро░рпН родрпЗро░рпНродро▓рпН 2019',
        part='snippet',
        type='video',
        regionCode='IN',#Region
        relevanceLanguage='ta',#Language
        maxResults=50  #
    ).execute()

    video_ids = []
    for search_result in search_response.get('items', []):
        video_ids.append(search_result['id']['videoId'])

#Get comments for each video_id retrived via api
    for video_id in video_ids:
        comment_response = yt.commentThreads().list(
            part='snippet',
            videoId=video_id,
            textFormat='plainText',
            maxResults=100 #number of comments per video
        ).execute()

        for comment_result in comment_response.get('items', []):
            comment = comment_result['snippet']['topLevelComment']['snippet']['textDisplay']
            print(comment)

except googleapiclient.errors.HttpError as e:
    print("An error occurred:", e)
```
